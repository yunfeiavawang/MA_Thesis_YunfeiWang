```python
def classify_douban_post(text):
    url = "https://api.siliconflow.cn/v1/chat/completions"

    prompt = f"""基于这篇豆瓣帖子，判断它是在“寻求帮助”还是“提供帮助”：
    - **寻求帮助 (0)**: 发帖人希望获得帮助。通常包括：
      - 说明自身情况（如学术背景、求职情况等）
      - 提出具体问题，表达不确定感
      - 可能有情绪化表达（如“求助”“跪求”“好焦虑”等）
      - 可能提前感谢回复者（如“感谢大家”“希望得到指点”）

    - **提供帮助 (1)**: 发帖人希望分享经验或帮助他人。通常包括：
      - 分享自身经历，如刚获得offer、完成某项学术/职业目标
      - 主动提供信息或经验，未直接寻求帮助
      - 欢迎他人提问（如“大家可以问我”“希望帮助更多人”）

    ### **范例**
    1. Text: "个人背景本人本科双非，竞赛保研211理工科,真诚！跪求大家指点！说什么都感激不尽！" → **Category: 0 (寻求帮助)**
    2. Text: "大家好，希望能帮助更多女性学者，也希望更多女性进学术圈。我目前在英国高校任教，今年收到不少博士陶瓷邮件，面试也有几个。感觉这方面大家疑惑还挺多，得到的信息良莠不齐。大家可以向我提问" → **Category: 1 (提供帮助)**

    ### **分类要求**
    请严格按照以下要求分类：
    - 只输出**0**（寻求帮助）或**1**（提供帮助）
    - 不输出任何额外文本

    Now classify the following text:
    {text}
    """

    payload = {
        "model": "deepseek-ai/DeepSeek-V3",
        "messages": [{"role": "user", "content": prompt}],
        "stream": False,
        "max_tokens": 10,
        "temperature": 0,
        "top_p": 1,
        "n": 1
    }

    headers = {
        "Authorization": "Bearer token",  # Replace with actual key
        "Content-Type": "application/json"
    }

    response = requests.post(url, json=payload, headers=headers)

    if response.status_code == 200:
        result = response.json()
        classification = result["choices"][0]["message"]["content"].strip()
        return int(classification) if classification in ["0", "1"] else None
    else:
        print("Error:", response.text)
        return None

# Example Usage
text_example = "请问有保研面试经验的朋友吗？我好焦虑，不知道怎么准备，求指导！"
result = classify_douban_post(text_example)
print("Classification:", result)
```

```python
import pandas as pd
import csv
import requests
import time

# Define API and CSV filename
csv_filename = "/content/drive/MyDrive/MA_Thesis/DBCommunity/saved_data/deepseek.csv"
url = "https://api.siliconflow.cn/v1/chat/completions"
headers = {
    "Authorization": "Bearer token",  # Replace with actual key
    "Content-Type": "application/json"
}

# # Extract "Text" column and truncate to 100 words
# def truncate_text(text, word_limit=100):
#     words = text.split()
#     return " ".join(words[:word_limit])

# df["Truncated_Text"] = df["Text"].apply(truncate_text)

failed_rows = []  # Track failed rows for retrying

# **Start processing from Row 142 onward**
for i, text in df.loc[142:, "Truncated_Text"].items():
    success = False
    max_retries = 3  # Allow 3 retries for rate-limited requests
    retry_delay = 60  # Wait time in seconds before retrying

    for attempt in range(max_retries):
        prompt = f"""基于这篇豆瓣帖子，判断它是在“寻求帮助”还是“提供帮助”：
        - **寻求帮助 (0)**: 发帖人希望获得帮助。通常包括：
          - 说明自身情况（如学术背景、求职情况等）
          - 提出具体问题，表达不确定感
          - 可能有情绪化表达（如“求助”“跪求”“好焦虑”等）
          - 可能提前感谢回复者（如“感谢大家”“希望得到指点”）

        - **提供帮助 (1)**: 发帖人希望分享经验或帮助他人。通常包括：
          - 分享自身经历，如刚获得offer、完成某项学术/职业目标
          - 主动提供信息或经验，未直接寻求帮助
          - 欢迎他人提问（如“大家可以问我”“希望帮助更多人”）

        ### **分类要求**
        - 只输出**0**（寻求帮助）或**1**（提供帮助）
        - 不输出任何额外文本

        Now classify the following text:
        {text}
        """

        payload = {
            "model": "deepseek-ai/DeepSeek-V3",
            "messages": [{"role": "user", "content": prompt}],
            "stream": False,
            "max_tokens": 10,
            "temperature": 0,
            "top_p": 1,
            "n": 1
        }

        response = requests.post(url, json=payload, headers=headers)

        if response.status_code == 200:
            result = response.json()
            category = result["choices"][0]["message"]["content"].strip()

            # Append new results to CSV
            with open(csv_filename, mode="a", newline="", encoding="utf-8") as file:
                writer = csv.writer(file)
                writer.writerow([i, category])

            print(f"Processed Row {i}")  # Print only index
            success = True
            break  # Exit retry loop if successful

        elif "TPM limit reached" in response.text:
            print(f"Rate limit reached at Row {i}, retrying in {retry_delay} seconds...")
            time.sleep(retry_delay)  # Wait before retrying
        else:
            print(f"Error at Row {i}: {response.text}")
            break  # Break if it's a non-rate-limit error

    if not success:
        failed_rows.append(i)
        print(f"Failed to process Row {i} after {max_retries} retries.")

    time.sleep(3)  # Prevent exceeding rate limits

# **Retry failed rows separately**
if failed_rows:
    print("\nRetrying failed rows...")
    for i in failed_rows:
        text = df.loc[i, "Truncated_Text"]
        success = False

        for attempt in range(max_retries):
            payload["messages"][0]["content"] = f"Now classify the following text:\n{text}"
            response = requests.post(url, json=payload, headers=headers)

            if response.status_code == 200:
                result = response.json()
                category = result["choices"][0]["message"]["content"].strip()

                with open(csv_filename, mode="a", newline="", encoding="utf-8") as file:
                    writer = csv.writer(file)
                    writer.writerow([i, category])

                print(f"Retried Row {i} successfully.")
                success = True
                break
            else:
                print(f"Retry Error at Row {i}: {response.text}")
                time.sleep(retry_delay)

        if not success:
            print(f"Final failure at Row {i}, unable to process.")
```
